# Class

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def handsome_level(self):
        if (self.age > 20):
          return self.name + ' with ' + str(self.age) + ' years old is handsome' 
        else:
          return self.name + ' with ' + str(self.age) + ' years old is ugly' 
```

```python
st = Student("Tín",25)

print(st.name)
print(st.age)
print(st.handsome_level())
```

```python
class Tokenizer:
    def __init__(self):
        self._tokenized_list = []
    
    def tokenize(self, s):
        self._tokenized_list = s.split()
        return self._tokenized_list
    
    @property
    def list_token(self):
        return self._tokenized_list
    
    @property
    def number_of_words(self):
        return len(self._tokenized_list)
```

```python
tokenizer = Tokenizer()

s = "Statistical Machine Learning"
tokenizer.tokenize(s)
```

```python
lst_word = tokenizer.list_token
num_word = tokenizer.number_of_words
```
